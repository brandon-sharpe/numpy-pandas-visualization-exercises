## 4

Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
from env import host, user, password
def get_db_url(database):
    url = f'mysql+pymysql://{user}:{password}@{host}/{database}'
    return url
```

```python
chipotle = pd.read_sql('SELECT * FROM orders', get_db_url('chipotle'))
chipotle
```

```python
chipotle.item_price = chipotle.item_price.str.replace(r'$', '').astype(float)
```

```python
most_popular = chipotle[['item_name', 'quantity','item_price']].groupby(by = 'item_name').sum().nlargest(4, 'quantity', keep='all')
most_popular['item_name'] = most_popular.index
```

```python
sns.barplot(x='quantity',y='item_price', hue = 'item_name', data=most_popular)
```
