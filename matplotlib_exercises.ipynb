## 3

Combine the figures you created in the last step into one large figure with 4 subplots.

```python
plt.figure(figsize=(14, 6))
x = list(range(1,10))

y1 = [(n ** .5) for n in x]
y2 = [(n ** 3) for n in x]
y3 = [(2 ** n) for n in x]
y4 = [(1/ (n + 1)) for n in x]
y_list = [y1, y2, y3, y4]
titles = ['$y = \u221Ax$', '$y = x^3$', '$y = 2^x$', '$y = \dfrac{1}{(x+1)}$']

for i, y_val in enumerate(y_list):
    plt.subplot(1, 4, i+1)
    plt.plot(x, y_val)
    plt.title(titles[i])
```
