## 1.

Copy the users and roles DataFrames from the examples above

```python
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
```

```python
users
```

```python
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
```

```python
roles
```

## 2. 

What is the result of using a right join on the DataFrames?

```python
(users.merge(roles, 
            left_on='role_id', 
            right_on='id', 
            how='right')
    .drop(columns='role_id')
    .rename(columns={'id_x': 'id', 
                     'name_x': 'employee',
                     'id_y': 'role_id',
                     'name_y': 'role'}
            )
)
```

## 3.

What is the result of using an outer join on the DataFrames?

```python
(users.merge(roles, 
            left_on='role_id', 
            right_on='id', 
            how='outer')
    .drop(columns='role_id')
    .rename(columns={'id_x': 'id', 
                     'name_x': 'employee',
                     'id_y': 'role_id',
                     'name_y': 'role'}
            )
)
```

## 4.

What happens if you drop the foreign keys from the DataFrames and try to merge them?

```python
(users.merge(roles, 
            left_on='role_id', 
            right_on='id'
            how='outer')
    .drop(columns='id_x')
    .rename(columns={'id_x': 'id', 
                     'name_x': 'employee',
                     'name_y': 'role'}
            )
)
```

## 12.

Create a column named mileage_difference like you did in the DataFrames exercises; this column should contain the difference between highway and city mileage for each car.

```python
mpg_1 = mpg.assign(mileage_difference = mpg.hwy - mpg.cty)
mpg_1 
```

## 13. 

Create a column named average_mileage like you did in the DataFrames exercises; this is the mean of the city and highway mileage.

```python
mpg_1 = mpg_1.assign(average_mileage = (mpg.hwy+mpg.cty)/2)
mpg_1
```

## 14. 

Create a new column on the mpg dataset named is_automatic that holds boolean values denoting whether the car has an automatic transmission.

```python
mpg = mpg_1.assign(is_automatic = np.where( mpg.trans.str.startswith('auto') , 'Auto', 'Manual'))
mpg
```

## 15.

Using the mpg dataset, find out which which manufacturer has the best miles per gallon on average?

```python
mpg.groupby('manufacturer').average_mileage.agg('mean').nlargest(1, keep = 'all')
```

## 16. 

Do automatic or manual cars have better miles per gallon?

```python
mpg.groupby('is_automatic').average_mileage.mean()
```

## 7

Write the code necessary to create a cross tabulation of the number of titles by department. (Hint: this will involve a combination of SQL code to pull the necessary data and python/pandas code to perform the manipulations.)

```python
employees_dept = pd.read_sql('''SELECT * FROM titles
                JOIN dept_emp 
                ON titles.emp_no = dept_emp.emp_no
                JOIN departments
                ON dept_emp.dept_no = departments.dept_no''',
                get_db_url('employees')
    )
```

```python
employees_dept.columns=['emp_no','title','from_date','to_date','trash1','dept_no','trash2','trash3','trash4','dept_name']
```

```python
employees_dept.drop(['trash1','trash2','trash3','trash4'], axis=1)
```

```python
pd.crosstab(employees_dept.title, employees_dept.dept_name)
```
